```python
from transformers import pipeline
import re
```

```python
def clean_text(text):
    text = str(text)
    text = text.replace("<br />","")
    text = re.sub(r"@[A-Za-z0-9]+", ' ', text)
    text = re.sub(r"https?://[A-Za-z0-9./]+", ' ', text)
    text = re.sub(r"[^a-zA-z.!?'0-9]", ' ', text)
    text = re.sub('\t', ' ',  text)
    text = re.sub(r'[0-9]+', '', text)

    return text
```

```python
model_name = "textattack/xlnet-base-cased-imdb" #You can choose other models from https://huggingface.co/
sentimentanalyzer = pipeline("sentiment-analysis", model=model_name, tokenizer=model_name)
```

```python
text = clean_text("Ok the movie seem's interesting but that is all I'll give it this guy is just terrible at movies him Michael Baye James Wang Roland Emmeric need to all be fired.")
result = sentimentanalyzer(text)[0]
print(f"Sentiment: {result['label']}",result["score"])

text = " "
result = sentimentanalyzer(text)[0]
print(f"Sentiment: {result['label']}", result['score'])

text = "This is good"
result = sentimentanalyzer(text)[0]
print(f"Sentiment: {result['label']}", result['score'])
```
